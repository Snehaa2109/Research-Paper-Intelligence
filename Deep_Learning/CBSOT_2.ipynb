```python
#creating embeddings using sentence transformer model
import os
import numpy as np
if os.path.exists("paper_embeddings.npy"):
  print("Loading saved embeddings file in loaded_emb")
  loaded_emb = np.load('paper_embeddings.npy')
else:
  print("Generating embeddings")
  embeddings = model.encode(df["paper_text"].to_list(), batch_size = 32, show_progress_bar = True)
  np.save('paper_embeddings.npy', embeddings)
```

```python
#creating index on the dimensionality of each vector embedding
import os
import numpy as np
import faiss
if os.path.exists("faiss_index.npy"):
    print("loading the saved index file")
    index = faiss.read_index('faiss_index.npy')
else:
    print("Generating the index file")
    index = faiss.IndexFlatIP(loaded_emb.shape[1])
    index.add(loaded_emb)
    faiss.write_index(index,'faiss_index.npy')
```

```python
query = ['Transformers in Deep Learning and its self attention mechanism']
query_emb = model.encode(query)
type(query_emb)
query_emb.shape
```

```python
# Normalizing query embeddings
faiss.normalize_L2(query_emb)
type(query_emb)
```

```python
#performing search via faiss(which uses cosine similarity)
D,I = index.search(query_emb,5)
D,I
```

## Main Function to Generate similarity score##

```python
# Creating a general function for finding similarity score of list of queries
def get_similarity_score (query,k=5):
    query_embed = model.encode(query)
    faiss.normalize_L2(query_embed)
    D,I = index.search(query_embed,k)
    return D,I
```

```python
get_similarity_score(query,6)
```
